```python
# Required Imports
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Initializing Variables
filename = 'p2.csv'
data = pd.read_csv(filename)
header = list(data.columns.values)
```

```python
col_mean = []
col_var = []

# Plotting Histograms
for h in header[:-1]:
    pd.DataFrame.hist(data[[h]])
    plt.title("Histogram for " + h)
    plt.xlabel(h)
    plt.ylabel("No. of Obs")
    plt.show()
    # plt_name = os.getcwd() + "/task1/histograms/hist_" + h
    # plt.savefig(plt_name)
    col_mean.append(pd.DataFrame.mean(data[[h]])[0])
    col_var.append(pd.DataFrame.var(data[[h]])[0])
```

The histograms for *X_i* are plotted above.

Histogram of **X_1** is normally distributed with fat tails.

Histogram of **X_2** tells us that there are 6 distinct bins for that variable and they all occur with nearly same frequency.

Histogram of **X_3** and **X_5** both demonstrate a decreasing exponential behavior.

Histogram of **X_4** is like **X_2**, with 3 distinct bins, and most values located between 0-1.

```python
print(header[:-1])
print("\nColumn Means")
print(col_mean)
print("\nColumn Variance")
print(col_var)
```

By looking at Histograms, Means, and Variances of Dependent Variables,
we observe the following: 

| Variable | Observation about Mean | Observation about Variance 
| :-------: | ---------------------- | -------------------------- 
| X1 | Near center of Histogram | High as values are equally widely spread out
| X2 | Approx center of Histogram | As similar number of obs in each bin, variance is comparable to mean
| X3 | Near left end of Histogram (values concentrated on left) | Low, as most obserations centered towards left
| X4 | Near approx center of Histogram | High value suggests possible outliers
| X5 | Near left end of Histogram | High value

```python
corr_mat = pd.DataFrame.corr(data, method='pearson')
print("\nCorrelation Matrix")
print(corr_mat)
```

```python
corr_mat[abs(corr_mat) > 0.5]
```

From the above table, which only shows entries whose absolute value is *> 0.5*,
we can see that - 

There is a positive correlation between variables **X1** and **X5** (value = 0.748).

Also, there is a negative correlation between variables **X2** and **X4** (value = -0.792).

So these variables might not be completely independent.

You can also see a positive correlation between *(Y, X1)* and *(Y, X5)*.
