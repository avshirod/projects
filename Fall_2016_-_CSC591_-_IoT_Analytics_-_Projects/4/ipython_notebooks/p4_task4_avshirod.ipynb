Using the *[adjusted_score metric](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.adjusted_rand_score.html#sklearn.metrics.adjusted_rand_score)*, we compare the cluster IDs created by all three methods.

This is the only way we can compare the method outcomes, because we don't know the *True* cluster IDs for given data.

We could have used inter-cluster and intra-cluster distance, if we were given the raw clusters.

Hence, in conclusion, for clustering methods - 
* Hierarchical Clustering with *'Average Linkage'*
* K-Means with *k=3*
* DBSCAN with *Eps=23* and *MinPts=6*

DBSCAN seems to fair better for given data.
